# Singular Value Decomposition

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
y1_df = pd.read_csv("./Y1.csv").T.to_numpy()
y2_df = pd.read_csv("./Y2.csv").T.to_numpy()
```

```python
c1 = np.cov(y1_df)
U_c1, S_c1, VT_c1 = np.linalg.svd(c1, full_matrices=False)
smat1 = np.diag(S_c1)
```

```python
c2 = np.cov(y2_df)
U_c2, S_c2, VT_c2 = np.linalg.svd(c2, full_matrices=False)
smat2 = np.diag(S_c2)
```

```python
M_a = np.dot(U_c1, (np.dot(np.sqrt(smat1), U_c2)))
M_a
```

```python
c_11 = np.dot(y1_df, y1_df.T)
U_11, S_11, VT_11 = np.linalg.svd(c_11, full_matrices=False)
smat_11 = np.diag(S_11)
```

```python
c_22 = np.dot(y2_df, y2_df.T)
U_22, S_22, VT_22 = np.linalg.svd(c_22, full_matrices=False)
smat_22 = np.diag(S_22)
```

```python
M_b = np.dot(U_11, np.dot(np.sqrt(smat_11) / 50000, U_22))
M_b
```

```python
mean = [0, 0]
cov = [[1, 0], [0, 1]]
x, y = np.random.multivariate_normal(mean, cov, 50000).T
```

```python
plt.plot(x, y, 'x')
plt.axis('equal')
plt.title('X')
plt.show()
```

```python
plt.plot(np.dot(M_a, [x, y])[0], np.dot(M_a, [x, y])[1], 'x', label='Y1\'')
plt.plot(y1_df[0], y1_df[1], 'x', label='Y1')
plt.axis('equal')
plt.legend()
plt.show()
```

```python
plt.plot(np.dot(M_a.T, [x, y])[0], np.dot(M_a.T, [x, y])[1], 'x', label='Y2\'')
plt.plot(y2_df[0], y2_df[1], 'x', label='Y2')
plt.axis('equal')
plt.legend()
plt.show()
```

```python
plt.plot(np.dot(M_b, [x, y])[0], np.dot(M_b, [x, y])[1], 'x', label='Y1\'')
plt.plot(y1_df[0], y1_df[1], 'x', label='Y1')
plt.axis('equal')
plt.legend()
plt.show()
```

```python
plt.plot(np.dot(M_b.T, [x, y])[0], np.dot(M_b.T, [x, y])[1], 'x', label='Y2\'')
plt.plot(y2_df[0], y2_df[1], 'x', label='Y2')
plt.axis('equal')
plt.legend()
plt.show()
```
